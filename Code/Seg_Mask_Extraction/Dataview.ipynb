```python
transform = transforms.Compose([transforms.Resize((500,500)),transforms.ToTensor()])
dataset = VOCDataSet(transform=transform)
dataloader = DataLoader(dataset, batch_size=16, shuffle=True)
```

```python
images, masks, index = iter(dataloader).next()

fig,axs= plt.subplots(4,4,figsize=(10,10),sharey=True,sharex=True,dpi=100)

cmap_green2trans = ListedColormap(['green',(0,0,0,0)])

for id,ax in enumerate(axs.flat):
    ax.axis('off')
    img = images[id]
    mask = masks[id]
    ax.imshow(np.transpose(img,(1,2,0)))
    ax.imshow(np.transpose(mask,(1,2,0)), alpha=0.75, cmap=cmap_green2trans)
```
